### 1. Diagnoses of substance use problems in grade 9 to 11 in the 2019 dataset, sorted in descending order
> With a total count of 2835, the most reported issue is **Cannabis dependence** (1066), followed by **Other mental health condition** (961) and **Cannabis abuse** (436). This underscores the importance of implementing targeted interventions and support systems to address cannnabis use problems in grade 9 to 11.

```python
dsmcrit_mapping = {
    1: 'Alcohol-induced disorder',
    2: 'Substance-induced disorder',
    3: 'Alcohol intoxication',
    4: 'Alcohol dependence',
    5: 'Opioid dependence',
    6: 'Cocaine dependence',
    7: 'Cannabis dependence',
    8: 'Other substance dependence',
    9: 'Alcohol abuse',
    10: 'Cannabis abuse',
    11: 'Other substance abuse',
    12: 'Opioid abuse',
    13: 'Cocaine abuse',
    14: 'Anxiety disorders',
    15: 'Depressive disorders',
    16: 'Schizophrenia/other psychotic disorders',
    17: 'Bipolar disorders',
    18: 'Attention deficit/disruptive behavior disorders',
    19: 'Other mental health condition',
    -9: 'Missing/unknown/not collected/invalid/no or deferred diagnosis'
}

counts_dsmcrit_g12_2019 = df_2019_clean[df_2019_clean['EDUC'] == 2]['DSMCRIT'].value_counts()
counts_dsmcrit_g12_2019.index = counts_dsmcrit_g12_2019.index.map(dsmcrit_mapping)
counts_dsmcrit_g12_2019_sorted = counts_dsmcrit_g12_2019.sort_values(ascending=False)

print(counts_dsmcrit_g12_2019_sorted)
print("Total counts:", counts_dsmcrit_g12_2019.sum())
```

### 2. Dynamics of addiction treatment sector between 2019 and 2021
> Non-intensive outpatient services decreased by 5.16%, indicating possible changes in patient needs or treatment approaches. Long-term residential programs saw a considerable decrease of 7.51%, while short-term residential services witnessed a modest increase of 5.87%, possibly reflecting shifts in treatment duration preferences or service availability.

```python
services_mapping = {
    1: 'Detox, 24-hour, hospital inpatient',
    2: 'Detox, 24-hour, free-standing residential',
    3: 'Rehab/residential, hospital (non-detox)',
    4: 'Rehab/residential, short term (30 days or fewer)',
    5: 'Rehab/residential, long term (more than 30 days)',
    6: 'Ambulatory, intensive outpatient',
    7: 'Ambulatory, non-intensive outpatient',
    8: 'Ambulatory, detoxification'
}

counts_services_2021 = df_2021_clean['SERVICES'].value_counts()
counts_services_2021.index = counts_services_2021.index.map(services_mapping)
counts_2021 = counts_services_2021.sum()
percentages_services_2021 = round((counts_services_2021 / counts_2021) * 100, 2)


counts_services_2019 = df_2019_clean['SERVICES'].value_counts()
counts_services_2019.index = counts_services_2019.index.map(services_mapping)
counts_2019 = counts_services_2019.sum()
percentages_services_2019 = round((counts_services_2019 / counts_2019) * 100, 2)

services_difference_percentage = percentages_services_2021.subtract(percentages_services_2019, fill_value=0)

psource_data = pd.DataFrame({
    '2019': counts_services_2019,
    '2021': counts_services_2021,
    '2019 %': percentages_services_2019,
    '2021 %': percentages_services_2021,
    'Difference %': services_difference_percentage
})

print(psource_data)
```

### 3. Referral source patterns reflected in the 2019 and 2021 datasets
> There is a significant decrease in referrals from **court/criminal justice referral/DUI/DWI**, alongside a substantial increase in referrals from **individual**. These shifts may reflect evolving societal dynamics, including policy changes concerning court referrals for young adults, and increased awareness about substance use problems and improved service availability or accessibility, particularly for self-referral.

```python
psource_mapping = {
    1: 'Individual (includes self-referral)',
    2: 'Alcohol/drug use care provider',
    3: 'Other health care provider',
    4: 'School (educational)',
    5: 'Employer/EAP',
    6: 'Other community referral',
    7: 'Court/criminal justice referral/DUI/DWI',
    -9: 'Missing/unknown/not collected/invalid'
}

counts_psource_2021 = df_2021_clean['PSOURCE'].value_counts()
counts_psource_2021.index = counts_psource_2021.index.map(psource_mapping)
percentages_psource_2021 = round((counts_psource_2021 / counts_2021) * 100, 2)


counts_psource_2019 = df_2019_clean['PSOURCE'].value_counts()
counts_psource_2019.index = counts_psource_2019.index.map(psource_mapping)
percentages_psource_2019 = round((counts_psource_2019 / counts_2019) * 100, 2)

psource_difference_percentage = percentages_psource_2021.subtract(percentages_psource_2019, fill_value=0)

psource_data = pd.DataFrame({
    '2019': counts_psource_2019,
    '2021': counts_psource_2021,
    '2019 %': percentages_psource_2019,
    '2021 %': percentages_psource_2021,
    'Difference %': psource_difference_percentage
})

print(psource_data)
```

### 4. Frequency of primary substance use between 2019 and 2021
> From the data, one could say the addictive level of some substances increase, or some subtances become more common among students. **Heroin**, **non-prescription methadone**, **other opiates and synthetics**, **other amphetamines**, and many other increased from **no use in the past month** to **daily use**. These changes suggest potential alterations in drug use behaviors and could inform targeted interventions and public health policies aimed at addressing new substances prevalent among young adults and their associated issues.

```python
sub_mapping = {
    1: 'None',
    2: 'Alcohol',
    3: 'Cocaine/crack',
    4: 'Marijuana/hashish',
    5: 'Heroin',
    6: 'Non-prescription methadone',
    7: 'Other opiates and synthetics',
    8: 'PCP',
    9: 'Hallucinogens',
    10: 'Methamphetamine/speed',
    11: 'Other amphetamines',
    12: 'Other stimulants',
    13: 'Benzodiazepines',
    14: 'Other tranquilizers',
    15: 'Barbiturates',
    16: 'Other sedatives or hypnotics',
    17: 'Inhalants',
    18: 'Over-the-counter medications',
    19: 'Other drugs',
    -9: 'Missing/unknown/not collected/invalid'
}

route_mapping = {
    1: 'Oral',
    2: 'Smoking',
    3: 'Inhalation',
    4: 'Injection (intravenous, intramuscular, intradermal, or subcutaneous)',
    5: 'Other',
    -9: 'Missing/unknown/not collected/invalid'
}

freq_mapping = {
    1: 'No use in the past month',
    2: 'Some use',
    3: 'Daily use',
    -9: 'Missing/unknown/not collected/invalid'
}
```

```python
most_frequent_2021 = df_2021_clean.groupby('SUB1').agg({'FREQ1': lambda x: x.mode()[0]})

most_frequent_2021.index = most_frequent_2021.index.map(sub_mapping)
most_frequent_2021_sorted = most_frequent_2021.sort_values(by='FREQ1', ascending=False)
most_frequent_2021_sorted['FREQ1'] = most_frequent_2021_sorted['FREQ1'].map(freq_mapping)


print(most_frequent_2021_sorted)
```

```python
most_frequent_2019 = df_2019_clean.groupby('SUB1').agg({'FREQ1': lambda x: x.mode()[0]})

most_frequent_2019.index = most_frequent_2019.index.map(sub_mapping)
most_frequent_2019_sorted = most_frequent_2019.sort_values(by='FREQ1', ascending=False)
most_frequent_2019_sorted['FREQ1'] = most_frequent_2019_sorted['FREQ1'].map(freq_mapping)


print(most_frequent_2019_sorted)
```

### 5.  Trends of substance use from 2019 to 2021
> In 2019 and 2021, **marijuana/hashish**, **alcohol**, and **cocaine/crack** were the most commonly reported substances. This data can assist in allocating public health efforts and initiating intervention strategies targeted at substances gaining prevalence, such as hallucinogens.

```python
sub_counts_2021 = {}

for column in ['SUB1', 'SUB2', 'SUB3']:
    counts = df_2021_clean[column].value_counts()
    sub_counts_2021[column] = counts

sub_counts_2021_df = pd.DataFrame(sub_counts_2021)

sub_counts_2021_df = sub_counts_2021_df.fillna(0).astype(int)

sub_counts_2021_df['Total'] = sub_counts_2021_df.sum(axis=1)

sub_counts_2021_df_sorted = sub_counts_2021_df.sort_values(by='Total', ascending=False)
sub_counts_2021_df_sorted.index = sub_counts_2021_df_sorted.index.map(sub_mapping)
sub_counts_2021_df_sorted.drop(index=['None'], inplace=True)

print(sub_counts_2021_df_sorted)
```

```python
sub_counts_2019 = {}

for column in ['SUB1', 'SUB2', 'SUB3']:
    counts = df_2019_clean[column].value_counts()
    sub_counts_2019[column] = counts

sub_counts_2019_df = pd.DataFrame(sub_counts_2019)

sub_counts_2019_df = sub_counts_2019_df.fillna(0).astype(int)

sub_counts_2019_df['Total'] = sub_counts_2019_df.sum(axis=1)

sub_counts_2019_df_sorted = sub_counts_2019_df.sort_values(by='Total', ascending=False)
sub_counts_2019_df_sorted.index = sub_counts_2019_df_sorted.index.map(sub_mapping)
sub_counts_2019_df_sorted.drop(index=['None'], inplace=True)

print(sub_counts_2019_df_sorted)
```
