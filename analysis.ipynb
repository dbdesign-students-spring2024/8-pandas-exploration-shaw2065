1. Diagnoses of substance use problems in grade 12 in the 2019 dataset, sorted in descending order.
- With a total count of 447, the most reported issue is **Other mental health condition** (234), followed by **Cannabis dependence** (65) and **Opioid dependence** (41). This underscores the importance of implementing targeted interventions and support systems to address substance use problems, especially the need to include other conditions apart from substances widely recognized as addictive, such as cannabis and opioids.

```python
dsmcrit_mapping = {
    1: 'Alcohol-induced disorder',
    2: 'Substance-induced disorder',
    3: 'Alcohol intoxication',
    4: 'Alcohol dependence',
    5: 'Opioid dependence',
    6: 'Cocaine dependence',
    7: 'Cannabis dependence',
    8: 'Other substance dependence',
    9: 'Alcohol abuse',
    10: 'Cannabis abuse',
    11: 'Other substance abuse',
    12: 'Opioid abuse',
    13: 'Cocaine abuse',
    14: 'Anxiety disorders',
    15: 'Depressive disorders',
    16: 'Schizophrenia/other psychotic disorders',
    17: 'Bipolar disorders',
    18: 'Attention deficit/disruptive behavior disorders',
    19: 'Other mental health condition',
    -9: 'Missing/unknown/not collected/invalid/no or deferred diagnosis'
}

counts_dsmcrit_g12_2019 = df_2019_clean[df_2019_clean['EDUC'] == 3]['DSMCRIT'].value_counts().sort_values()
counts_dsmcrit_g12_2019.index = counts_dsmcrit_g12_2019.index.map(dsmcrit_mapping)
counts_dsmcrit_g12_2019 = counts_dsmcrit_g12_2019.sort_values(ascending=False)

print(counts_dsmcrit_g12_2019)
print("Total counts:", counts_dsmcrit_g12_2019.sum())
```

2. Referral source patterns reflected in the 2019 and 2021 datasets, displaying both numerical differences and percentage differences
- The data suggests changes in referral pattern. There is a decrease in referrals from court/criminal justice systems and other community sources, alongside an increase in referrals from other healthcare providers. These shifts may reflect evolving societal dynamics, including increased awareness among healthcare providers beyond alcohol/drug care, policy adaptations concerning court referrals for young adults, and changes in service availability or accessibility, particularly self-referral.

```python
psource_mapping = {
    1: 'Individual (includes self-referral)',
    2: 'Alcohol/drug use care provider',
    3: 'Other health care provider',
    4: 'School (educational)',
    5: 'Employer/EAP',
    6: 'Other community referral',
    7: 'Court/criminal justice referral/DUI/DWI',
    -9: 'Missing/unknown/not collected/invalid'
}

counts_psource_2021 = df_2021_clean['PSOURCE'].value_counts()
counts_psource_2021.index = counts_psource_2021.index.map(psource_mapping)
counts_2021 = counts_psource_2021.sum()
percentages_psource_2021 = round((counts_psource_2021 / counts_2021) * 100, 2)


counts_psource_2019 = df_2019_clean['PSOURCE'].value_counts()
counts_psource_2019.index = counts_psource_2019.index.map(psource_mapping)
counts_2019 = counts_psource_2019.sum()
percentages_psource_2019 = round((counts_psource_2019 / counts_2019) * 100, 2)

psource_difference = counts_psource_2021.subtract(counts_psource_2019, fill_value=0).astype(int)
psource_difference_percentage = percentages_psource_2021.subtract(percentages_psource_2019, fill_value=0)

psource_data = pd.DataFrame({
    '2019': counts_psource_2019,
    '2021': counts_psource_2021,
    'Difference': psource_difference,
    '2019 %': percentages_psource_2019,
    '2021 %': percentages_psource_2021,
    'Difference %': psource_difference_percentage
})

print(psource_data)
```

3. Frequency of primary substance use between 2019 and 2021
- The data comparing substance use frequencies between 2019 and 2021 reveals shifts in consumption patterns. Some substances decreased in use, like "PCP", "Hallucinogens", "Cocaine/crack", and "Heroin", while others, such as "Benzodiazepines" and "Inhalants", increased. These changes suggest potential alterations in drug use behaviors, and could inform targeted interventions and public health policies aimed at addressing specific substances experiencing shifts in consumption patterns.

```python
sub_mapping = {
    1: 'None',
    2: 'Alcohol',
    3: 'Cocaine/crack',
    4: 'Marijuana/hashish',
    5: 'Heroin',
    6: 'Non-prescription methadone',
    7: 'Other opiates and synthetics',
    8: 'PCP',
    9: 'Hallucinogens',
    10: 'Methamphetamine/speed',
    11: 'Other amphetamines',
    12: 'Other stimulants',
    13: 'Benzodiazepines',
    14: 'Other tranquilizers',
    15: 'Barbiturates',
    16: 'Other sedatives or hypnotics',
    17: 'Inhalants',
    18: 'Over-the-counter medications',
    19: 'Other drugs',
    -9: 'Missing/unknown/not collected/invalid'
}

route_mapping = {
    1: 'Oral',
    2: 'Smoking',
    3: 'Inhalation',
    4: 'Injection (intravenous, intramuscular, intradermal, or subcutaneous)',
    5: 'Other',
    -9: 'Missing/unknown/not collected/invalid'
}

freq_mapping = {
    1: 'No use in the past month',
    2: 'Some use',
    3: 'Daily use',
    -9: 'Missing/unknown/not collected/invalid'
}
```

```python
most_frequent_2021 = df_2021_clean.groupby('SUB1').agg({'FREQ1': lambda x: x.mode()[0]})

most_frequent_2021.index = most_frequent_2021.index.map(sub_mapping)
most_frequent_2021_sorted = most_frequent_2021.sort_values(by='FREQ1', ascending=False)
most_frequent_2021_sorted['FREQ1'] = most_frequent_2021_sorted['FREQ1'].map(freq_mapping)


print(most_frequent_2021_sorted)
```

```python
most_frequent_2019 = df_2019_clean.groupby('SUB1').agg({'FREQ1': lambda x: x.mode()[0]})

most_frequent_2019.index = most_frequent_2019.index.map(sub_mapping)
most_frequent_2019_sorted = most_frequent_2019.sort_values(by='FREQ1', ascending=False)
most_frequent_2019_sorted['FREQ1'] = most_frequent_2019_sorted['FREQ1'].map(freq_mapping)


print(most_frequent_2019_sorted)
```

4.  Trends of substance use from 2019 to 2021
- In 2019 and 2021, marijuana/hashish, alcohol, cocaine/crack, and benzodiazepines were the most commonly reported substances. Heroin shows a decrease from 2019 to 2021, while hallucinogens exhibit an increase. This data can assist in allocating public health efforts and initiating intervention strategies targeted at substances gaining prevalence, such as hallucinogens.

```python
sub_counts_2021 = {}

for column in ['SUB1', 'SUB2', 'SUB3']:
    counts = df_2021_clean[column].value_counts()
    sub_counts_2021[column] = counts

sub_counts_2021_df = pd.DataFrame(sub_counts_2021)

sub_counts_2021_df = sub_counts_2021_df.fillna(0).astype(int)

sub_counts_2021_df['Total'] = sub_counts_2021_df.sum(axis=1)

sub_counts_2021_df_sorted = sub_counts_2021_df.sort_values(by='Total', ascending=False)
sub_counts_2021_df_sorted.index = sub_counts_2021_df_sorted.index.map(sub_mapping)
sub_counts_2021_df_sorted.drop(index=['None'], inplace=True)

print(sub_counts_2021_df_sorted)
```

```python
sub_counts_2019 = {}

for column in ['SUB1', 'SUB2', 'SUB3']:
    counts = df_2019_clean[column].value_counts()
    sub_counts_2019[column] = counts

sub_counts_2019_df = pd.DataFrame(sub_counts_2019)

sub_counts_2019_df = sub_counts_2019_df.fillna(0).astype(int)

sub_counts_2019_df['Total'] = sub_counts_2019_df.sum(axis=1)

sub_counts_2019_df_sorted = sub_counts_2019_df.sort_values(by='Total', ascending=False)
sub_counts_2019_df_sorted.index = sub_counts_2019_df_sorted.index.map(sub_mapping)
sub_counts_2019_df_sorted.drop(index=['None'], inplace=True)

print(sub_counts_2019_df_sorted)
```
